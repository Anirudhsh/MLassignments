```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 
```

```
'''data_visualization: 3 

use this data from pandas

i)  data source  http://13.234.66.67/summer19/datasets/bank.csv
ii)  read this data by pandas
iii)  plot all the possible graphs
iv) use all possible combination of  all columns
'''
```

```
ff=pd.read_csv('http://13.234.66.67/summer19/datasets/bank.csv')
```

```
print(ff.head(3))
```

```
#Estimated mean salaries of different countries
conts=ff['Geography'].values
fgg=ff.groupby('Geography')
ffg=fgg.agg(np.mean)
conts=list(fgg.groups.keys())
```

```
plt.figure(figsize=(10,10))
plt.pie(ffg['EstimatedSalary'],labels=conts)
```

```
plt.figure(figsize=(15,15))
plt.scatter(ff.Balance.values,ff.EstimatedSalary.values)
plt.grid(color='red')
```

```
#avg age of male and female customers of bank
ffk=ff.groupby('Gender')
mfa=ffk.agg(np.mean)
hemf=list(ffk.groups.keys())
hemf
```

```
plt.bar(hemf,mfa['Age'])
plt.bar(hemf,mfa['Tenure'])
```

```

plt.bar(hemf,mfa['EstimatedSalary'])
plt.bar(hemf,mfa['Balance'])
```

```
ffk=ff.groupby('Gender')
mfa=ffk.agg(np.sum)
hemf=list(ffk.groups.keys())
hemf
```

```
#no of males and females possesing credit cards
plt.bar(hemf,mfa['HasCrCard'])
```
