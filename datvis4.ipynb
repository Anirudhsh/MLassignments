```
from bs4 import BeautifulSoup
import requests
URL = "https://en.wikipedia.org/wiki/Machine_learning"
r = requests.get(URL) 
soup = BeautifulSoup(r.content, 'html5lib')
sdiv = soup.find("div", {"id": "mw-content-text"})
stext = sdiv.text
words=str(stext).replace("\n"," ")
words=words.split(" ")
wordcnt={}
for i in words:
    if i not in wordcnt:
        wordcnt[i]=1
    else:
        wordcnt[i]+=1
wosorted = sorted(wordcnt.items(), key =lambda x:(x[1], x[0]),reverse=True)
#print(wosorted)
# 0 not taken as it is ''
top20words = wosorted[1:21]
val3=[]
lab3=[]
for i in wosorted:
  if i[1]>3:
    val3.append(i[1])
    lab3.append(i[0])
  else:
    break
print(top20words)
```

```
import matplotlib.pyplot as plt
```

```
labs=[]
vals=[]
for i in top20words:
  vals.append(i[1])
  labs.append(i[0])
```

```
plt.figure(figsize=(15,15))
plt.pie(vals,labels=labs)
```

```
plt.figure(figsize=(15,15))
plt.bar(labs,vals)
```

```
plt.figure(figsize=(300,50))
plt.scatter(lab3,val3)
```
